```python
import os

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style('darkgrid')
```

```python
DATASET_DIR = './datasets/raw_data'
FIGURE_DIR = './figures'
```

```python
classes = os.listdir(DATASET_DIR)
num_of_classes = len(classes)
classes
```

```python
dataset_count_dict = {
    'Brand': [],
    'Number of Samples': []
}

for class_ in classes:
    files = os.listdir(os.path.join(DATASET_DIR, class_))
    dataset_count_dict['Brand'].append(class_)
    dataset_count_dict['Number of Samples'].append(len(files))

dataset_count_df = pd.DataFrame(data=dataset_count_dict)
dataset_count_df.head()
```

```python
dataset_count_df['Number of Samples'].sum()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

sns.barplot(x=dataset_count_df['Brand'],
            y=dataset_count_df['Number of Samples'],
            ax=ax, label="Number of Samples from each Class")

ax.set_xlabel("Brands")
ax.set_ylabel("Number of Samples")
plt.xticks(rotation=45)

plt.savefig(f'{FIGURE_DIR}/dataset_count.png')
plt.show()
```
